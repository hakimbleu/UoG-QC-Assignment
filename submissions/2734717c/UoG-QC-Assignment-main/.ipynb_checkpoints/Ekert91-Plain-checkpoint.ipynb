```python
from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_gate
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Operator
from qiskit.quantum_info import Statevector
from qiskit import transpile 
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit import ControlledGate
import math 

# The aim of the assignment is to simulate the Ekert91 key distribution protocol.

# This notebook is for a simulation of the protocol without an attacker.

root3 = math.sqrt(3)
root2 = math.sqrt(2)

Z = [[1, 0],
     [0, -1]]
X = [[0, 1],
     [1, 0]]
W = [[1 / root2,  1 / root2],
     [1 / root2, -1 / root2]]
V = [[-1 / root2, 1 / root2],
     [1 / root2, 1 / root2]]

alice_observables = [X, W, Z]
bob_observables = [W, Z, V]
N = 7

def entangledPair():
    q = QuantumCircuit(2)
    q.h(0)
    q.cx(0,1)
    q.x(1)
    q.z(0)
    return q

def random():
    q = QuantumCircuit(2)
    random_matrix = [[ 1 / root3, - root2 / root3 ],
                     [ root2 / root3, 1 / root3 ]]
    q.unitary(random_matrix, [1])
    q.h(0)
    q.measure_all()

    backend = BasicSimulator()
    compiled = transpile(q, backend)
    job_sim = backend.run(compiled, shots=1)
    result_sim = job_sim.result() 
    counts = result_sim.get_counts(compiled)
    return list(counts.keys())[0]

def measurement_choice():
    choice = random()
   
    if choice[0] == '0':
        choice = 0
    else:
        choice = int(choice[0]) + int(choice[1])

    
    return choice



alice_choices = []
bob_choices = []
results = []
for i in range ((9 * N) // 2):
    q = entangledPair()

    alice_choice = measurement_choice()
    alice_choices.append(alice_choice + 1)
    alice_measurement = alice_observables[alice_choice]
    
    bob_choice = measurement_choice()
    bob_choices.append(bob_choice + 1)
    bob_measurement = bob_observables[bob_choice]
    
    q.unitary(alice_measurement, [0])
    q.unitary(bob_measurement, [1])
    q.x(1)
    q.measure_all()

    backend = BasicSimulator()
    compiled = transpile(q, backend)
    job_sim = backend.run(compiled, shots=1)
    result_sim = job_sim.result() 
    result = list(result_sim.get_counts(compiled).keys())[0]

    results.append(result)
    

random()
final_key = []

for i in range((9 * N) // 2):
    if alice_choices[i] == 2 and bob_choices[i] == 1:
        final_key.append(results[i][0])
    elif alice_choices[i] == 3 and bob_choices[i] == 2:
        final_key.append(results[i][0])

final_key_total += len(final_key)



```
