```python
from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_gate
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Operator
from qiskit.quantum_info import Statevector
from qiskit import transpile,QuantumCircuit
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit import ControlledGate
import math 

# The aim of the assignment is to simulate the Ekert91 key distribution protocol.

# This notebook is for a simulation of the protocol with an attacker, to demonstrate that the attacker can be detected.

# The aim of the assignment is to simulate the Ekert91 key distribution protocol
# with an attacker (Eve)

# number of key bits wanted
N = 100
iterations = int(9*N/2)

backend = BasicSimulator()

# Bell matrices
root2 = math.sqrt(2)
denom1 = math.sqrt(4 + 2*root2)
denom2 = math.sqrt(4 - 2*root2)

W_matrix = [[-1/denom1, (1+root2)/denom1],
[1/denom2, (root2-1)/denom2]]

V_matrix = [[1/denom1, (1+root2)/denom1],
            [-1/denom2, (root2-1)/denom2]]

#the entangled pair
def entangledPair():
    qc = QuantumCircuit(2,2)  
    qc.h(0)
    qc.cx(0,1)
    return qc

def runCircuit(qc):
    compiled = transpile(qc, backend)
    job = backend.run(compiled, shots=1)
    result = job.result()
    counts = result.get_counts(compiled)
    return list(counts.keys())[0]
    
def quantum_random_bit():
    qc = QuantumCircuit(1,1)
    qc.h(0)
    qc.measure(0,0)

    compiled = transpile(qc, backend)
    job = backend.run(compiled, shots=1)
    result = job.result()
    counts = result.get_counts()

    return int(list(counts.keys())[0])

def quantum_random_1_3():
    U = [[math.sqrt(1/3), -math.sqrt(2/3)],[math.sqrt(2/3),  math.sqrt(1/3)]]
    qc = QuantumCircuit(1,1)
    qc.unitary(U, [0])
    qc.measure(0,0)

    compiled = transpile(qc, backend)
    job = backend.run(compiled, shots=1)
    result = job.result()
    counts = result.get_counts()
    first = list(counts.keys())[0]

    if first == "0":
        return 1

    qc2 = QuantumCircuit(1,1)
    qc2.h(0)
    qc2.measure(0,0)
    compiled2 = transpile(qc2, backend)
    job2 = backend.run(compiled2, shots=1)
    result2 = job2.result()
    counts2 = result2.get_counts()
    second = list(counts2.keys())[0]

    if second == "0":
        return 2
    else:
        return 3

def convert(bit):
    return 1 if bit=="0" else -1

#stores the results
key=[]
A1B1=[]
A1B3=[]
A3B1=[]
A3B3=[]

for i in range(iterations):
    qc = entangledPair()

    # Eve intercepts Bob's qubit
    qc.measure(1,1)   
    qc.reset(1)
    eve_bit = quantum_random_bit()
    if eve_bit == 1:
        qc.x(1)

    aliceChoice = quantum_random_1_3()
    bobChoice = quantum_random_1_3()

    # Alice operators
    if aliceChoice == 1: # X
        qc.h(0)
    elif aliceChoice == 2: # W
        qc.unitary(W_matrix,[0])
    elif aliceChoice == 3: # Z
        pass

    # Bob operators
    if bobChoice == 1: # W
        qc.unitary(W_matrix,[1])
    elif bobChoice == 2: # Z
        pass
    elif bobChoice == 3: # V
        qc.unitary(V_matrix,[1])

    qc.measure(0,0)
    qc.measure(1,1)

    result = runCircuit(qc)

    a = convert(result[1])
    b = convert(result[0])
    
    # make key
    if aliceChoice == 2 and bobChoice == 1:
        key.append(-b)

    if aliceChoice == 3 and bobChoice == 2:
        key.append(-b)
    # the bell inequality calculation
    if aliceChoice==1 and bobChoice==1:
        A1B1.append(a*b)

    if aliceChoice==1 and bobChoice==3:
        A1B3.append(a*b)

    if aliceChoice==3 and bobChoice==1:
        A3B1.append(a*b)

    if aliceChoice==3 and bobChoice==3:
        A3B3.append(a*b)

def avg(lst):
    return sum(lst)/len(lst)

S = abs(avg(A1B1) - avg(A1B3) + avg(A3B1)+avg(A3B3))

print("Shared key:", key[:N])
print("Key length:", len(key[:N]))
print("Bell value S =", S)
```
