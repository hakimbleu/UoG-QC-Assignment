```python
from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_gate
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Operator
from qiskit.quantum_info import Statevector
from qiskit import transpile
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit import ControlledGate
import math

# The aim of the assignment is to simulate the Ekert91 key distribution protocol.

# This notebook is for a simulation of the protocol with an attacker, to demonstrate that the attacker can be detected.

# Final shared secret key between Alice and Bob
shared_key = []

# Desired length of the final key
N = 100
print("Alice and Bob want to create a shared key of size: "+ str(N))
#constants that will be used to make the matrices later on such as the transformation matrices and random
root3 = math.sqrt(3)
root2 = math.sqrt(2)

denom1 = math.sqrt(4 + 2*root2)
denom2 = math.sqrt(4 - 2*root2)

# W basis transformation matrix
# Corresponds to measuring the observable W = (X + Z)/root2
W_transform_matrix = [ [ -1 / denom1 , (1 + root2) / denom1 ],
                       [  1 / denom2 , (root2 - 1) / denom2 ] ]
W_transform = Operator(W_transform_matrix)

# V basis transformation matrix
# Corresponds to measuring the observable V = (X - Z)/root2
V_transform_matrix = [ [  1 / denom1 , (1 + root2) / denom1 ],
                       [ -1 / denom2 , (root2 - 1) / denom2 ] ]
V_transform = Operator(V_transform_matrix)

backend = BasicSimulator()

#this is for the desired 33% random chance generator using quantum states
h3_transform_matrix = [ [ 1 / root3 , root2 / root3 ],
                            [ root2 /root3 , -1 / root3 ] ]
h3_transform = Operator(h3_transform_matrix)



#this function chooses 1 out of the 3 bases randomly
#A qubit is prepared in the state: 1/root3|0> + root2/root3|1>
#if result is 0 returns choice A1 or B1 for Alice and Bob respectively
#else if result is 1, run a simulation by applying the hadamard gate to get equal chances of 2 and 3
#thus making this a 33% random generator
def random():
    cr = QuantumCircuit(1, 1)
    cr.unitary(h3_transform,[0])
    cr.measure(0, 0)
    qc_compiled = transpile(cr, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)
    if list(counts.keys())[0] == "0":
        return 1
    else:
        cr = QuantumCircuit(1, 1)
        cr.h(0)
        cr.measure(0, 0)
        qc_compiled = transpile(cr, backend)
        job_sim = backend.run(qc_compiled, shots=1)
        result_sim = job_sim.result()
        counts = result_sim.get_counts(qc_compiled)
        if list(counts.keys())[0] == "0":
            return 2
        else:
            return 3

# function definition for the entangled pair 1/root2(|01> - |10>)
# q1 is alice and q0 is bob, #qiskit ordering
def entangledPair():
    q = QuantumCircuit(2,2)
    q.h(1)
    #1/root2(|00> + |10>)
    q.x(0)
    #1/root2(|01> + |11>)
    q.cx(1,0)
    #1/root2(|01> + |10>)
    q.z(1)
    #1/root2(|01> - |10>)
    return q

#arrays to alice and bob to keep a track of their basis choices when measuring the qubits
aliceBasis = []
bobBasis = []
#arrays for alice and bob to keep a track of their results after they measured with their respective basis choice
aliceBits = []
bobBits = []


rounds = round((9*N)/2)
#the main protocol which runs 9*N/2 times
print("Alice and Bob are measuring with their basis choices and recording both the basis choices and the measurement results. Ongoing")
for i in range(rounds):
    c = entangledPair()

    ### different attacker scenarios (all one liners so uncomment what you want to test)
    c.measure(0,0) #attacker Eve interferes with Bobs qubit before Bob gets it
    #c.x(1) #attacker Eve negated Alice's qubit, changing the behaviour and it achieves near 0 in the bell inequality test

    
    # Alice randomly chooses measurement operator Ai
    # A1 = X, A2 = W, A3 = Z
    aliceC = random()
    if aliceC == 1:
      c.h(1)
    elif aliceC == 2:
      c.unitary(W_transform, [1])
    else:
      pass


    # Bob randomly chooses measurement operator Bj
    # B1 = W, B2 = Z, B3 = V
    bobC = random()
    if bobC == 1:
      c.unitary(W_transform, [0])
    elif bobC == 2:
      pass
    else:
      c.unitary(V_transform, [0])

    # alice and bob record the basis choices
    aliceBasis.append(aliceC)
    bobBasis.append(bobC)

    #measure
    c.measure_all()
    qc_compiled = transpile(c, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)
    result = list(counts.keys())[0]
    # Qiskit returns the classical bits as a string.
    # Here result[0] corresponds to Alice's qubit and result[1] to Bob's.
    aliceBits.append(result[0])
    bobBits.append(result[1])



# These measurement combinations are used for the CHSH Bell inequality test:
# (A1,B1), (A1,B3), (A3,B1), (A3,B3)
Slist = [[1,1],[1,3],[3,1],[3,3]]

# dictionaries for the different combinations which will be used to calculate the average when basis don't match
total = {"11": 0, "13":0 , "31":0 ,"33":0}
counts = {"11": 0, "13":0 , "31":0 ,"33":0}
print("Done")
print("Alice and Bob now compare their basis choices and try to make a shared key and also check the CHSH Bell Inequality test for any signs of eavesdroppers. Ongoing")
print("Sneaky attacker may interfere hehe")
#alice and bob now check their results after doing the measurements
for i in range(rounds):
  # when basis match it forms the final key
  if (aliceBasis[i] == 3 and bobBasis[i] == 2) or (aliceBasis[i] == 2 and bobBasis[i] == 1):
    # as agreed in the beginning that bob will invert his bits since the entangled state is correlated such that alice and bob get opposite results
    bobBits[i] = str(1 - int(bobBits[i]))

      
    #producing a key of length N as required so no need to append any further and keep going to calculate the CHSH bell inequality
    if len(shared_key) >= N:
      continue

    #append when basis match and when shared_key is still less than N
    shared_key.append(aliceBits[i])

  else:
    # if basis dont match then we check if the combination is part of the CHSH Bell Inequality test
    # if it is we record the results for the CHSH Bell Inequality test
    m = [aliceBasis[i], bobBasis[i]]
    if  m in Slist:
      aliceConvert = 1 if aliceBits[i] == "0" else -1
      bobConvert = 1 if bobBits[i] == "0" else -1
      if m == [1,1]:
        total["11"] += aliceConvert * bobConvert
        counts["11"]+=1

      if m == [1,3]:
        total["13"] += aliceConvert * bobConvert
        counts["13"]+=1

      if m == [3,1]:
        total["31"] += aliceConvert * bobConvert
        counts["31"]+=1

      if m == [3,3]:
        total["33"] += aliceConvert * bobConvert
        counts["33"]+=1

# Calculate the average for each combination of the measurements mentioned above
average = {"11": 0, "13":0 , "31":0 ,"33":0}
for k in total:
    if counts[k] != 0: #to avoid division by 0 error
        average[k] = total[k]/counts[k]

print("Done")

print("Finally, the results!")
# If the protocol works correctly:
# S ≈ 2√2 ≈ 2.828

# If an attacker interferes:
# S ≤ 2
# CHSH Bell Inequality Test
S = abs(average["11"] - average["13"] + average["31"] + average["33"])
print("Ideal S ≈ " + str(2*root2))
#result is always consistently less than 2 showing that alice and bob can always detect an attacker since attacker interfered with the entangled state
print("CHSH Bell inequality value S =", S) 
print("length of shared key is: "+ str(len(shared_key)))
```
